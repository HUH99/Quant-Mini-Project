# **Practice:** 금융상품 움직임 분석

```python
# 도구 불러오기
import pandas as pd
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Data Loading:** 분석에 사용할 데이터 불러오기

```python
# 파일 경로 구성
file_path = os.path.join('.', 'Data', 'IBMWatsonMarketing', 'Marketing_CustomerValue.csv')

# 데이터 로딩
data = pd.read_csv(file_path)

# 불필요한 열 제거
data_cleaned = data.drop(columns=['Effective To Date', 'Customer'])
```

## **Data Understanding:** 목표 달성을 위한 고객특징 이해하기

```python
# 보험 갱신 비율 계산
renewal_rate = data_cleaned['Response'].value_counts(normalize=True) * 100

# 개인정보에 따른 갱신 비율 분석
personal_info_columns = ['State', 'Education', 'EmploymentStatus', 'Gender', 'Location Code', 'Marital Status']
personal_info_renewal = pd.DataFrame()

for col in personal_info_columns:
    renewal_rate_by_col = data_cleaned.groupby(col)['Response'].value_counts(normalize=True).unstack() * 100
    personal_info_renewal = pd.concat([personal_info_renewal, renewal_rate_by_col], axis=1)

# 개인정보에 따른 갱신 비율 분석 및 시각화
plt.figure(figsize=(15, 10))

for i, col in enumerate(personal_info_columns, 1):
    plt.subplot(2, 3, i)
    renewal_rate_by_col = data_cleaned.groupby(col)['Response'].value_counts(normalize=True).unstack() * 100
    sns.barplot(x=renewal_rate_by_col.index, y=renewal_rate_by_col[1], palette='coolwarm')
    plt.title(f'Renewal Rate by {col}')
    plt.ylabel('Renewal Rate (%)')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 교육 수준, 거주지 유형, 혼인 여부에 따른 갱신 비율 분석
education_renewal_rate = data_cleaned.groupby('Education')['Response'].value_counts(normalize=True).unstack() * 100
location_code_renewal_rate = data_cleaned.groupby('Location Code')['Response'].value_counts(normalize=True).unstack() * 100
marital_status_renewal_rate = data_cleaned.groupby('Marital Status')['Response'].value_counts(normalize=True).unstack() * 100

# 결과 출력
renewal_rate, education_renewal_rate.loc[:, 1], location_code_renewal_rate.loc[:, 1], marital_status_renewal_rate.loc[:, 1]
```

보험 갱신 비율: 전체 응답자 중 약 14.32%가 보험을 갱신했습니다.

개인정보에 따른 갱신 비율:
교육 수준(Education): 'Doctor' 수준의 교육을 받은 고객이 약 17.54%로 가장 높은 갱신 비율을 보였습니다.

거주지 유형(Location Code): 'Suburban' 지역에 거주하는 고객이 약 17.44%로 가장 높은 갱신 비율을 보였습니다.

혼인 여부(Marital Status): 'Divorced' 상태인 고객이 약 23.67%로 가장 높은 갱신 비율을 보였습니다.

## **Data Understanding:** 목표 달성을 위한 보험특징 이해하기

```python
# 보험 조건에 따른 갱신 비율 분석 및 시각화
insurance_features = ['Number of Open Complaints', 'Number of Policies', 'Vehicle Class', 'Vehicle Size', 'Sales Channel']

plt.figure(figsize=(15, 10))

for i, feature in enumerate(insurance_features, 1):
    plt.subplot(2, 3, i)
    renewal_rate_by_feature = data_cleaned.groupby(feature)['Response'].value_counts(normalize=True).unstack() * 100
    sns.barplot(x=renewal_rate_by_feature.index, y=renewal_rate_by_feature[1], palette='coolwarm')
    plt.title(f'Renewal Rate by {feature}')
    plt.ylabel('Renewal Rate (%)')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 3개 이상의 보험증권을 가진 고객 중에서 가장 갱신 비율이 많은 보험증권의 수 파악
policies_renewal_rate = data_cleaned[data_cleaned['Number of Policies'] >= 3].groupby('Number of Policies')['Response'].value_counts(normalize=True).unstack() * 100
max_renewal_rate_policy_count = policies_renewal_rate[1].idxmax()

# 자동차 등급, 크기, 마케팅 채널에 따른 갱신 비율이 가장 높은 조건 파악
vehicle_class_max_renewal = data_cleaned.groupby('Vehicle Class')['Response'].value_counts(normalize=True).unstack() * 100
vehicle_size_max_renewal = data_cleaned.groupby('Vehicle Size')['Response'].value_counts(normalize=True).unstack() * 100
sales_channel_max_renewal = data_cleaned.groupby('Sales Channel')['Response'].value_counts(normalize=True).unstack() * 100

max_renewal_rate_vehicle_class = vehicle_class_max_renewal[1].idxmax()
max_renewal_rate_vehicle_size = vehicle_size_max_renewal[1].idxmax()
max_renewal_rate_sales_channel = sales_channel_max_renewal[1].idxmax()

max_renewal_rate_policy_count, max_renewal_rate_vehicle_class, max_renewal_rate_vehicle_size, max_renewal_rate_sales_channel
```

자동차 등급에 따른 가장 높은 갱신 비율: 'Sports Car' 등급의 자동차를 가진 고객들이 가장 높은 갱신 비율을 보였습니다.

자동차 크기에 따른 가장 높은 갱신 비율: 'Large' 크기의 자동차를 가진 고객들이 가장 높은 갱신 비율을 보였습니다.

마케팅 채널에 따른 가장 높은 갱신 비율: 'Agent' 채널을 통해 마케팅된 고객들이 가장 높은 갱신 비율을 보였습니다.


## **Business Strategy:** 이익 증가를 위한 전략 제시하기

```python
# 보험 청구 금액이 1000 이상인 고객 데이터 필터링
high_claims_data = data_cleaned[data_cleaned['Total Claim Amount'] >= 1000]

# 분석 대상 변수
analysis_variables = ['State', 'Education', 'EmploymentStatus', 'Gender', 'Marital Status', 'Vehicle Class', 'Vehicle Size', 'Sales Channel']

# 각 변수별 갱신 비율과 갱신하지 않는 비율의 차이 계산
renewal_diffs = {}

for var in analysis_variables:
    renewal_rates = high_claims_data.groupby(var)['Response'].value_counts(normalize=True).unstack() * 100
    renewal_diff = (renewal_rates[1] - renewal_rates[0]).abs()
    max_diff_var_value = renewal_diff.idxmax()
    max_diff_value = renewal_diff.max()
    renewal_diffs[var] = {'Max Diff Value': max_diff_var_value, 'Difference': max_diff_value}

# 가장 큰 차이를 보이는 변수와 값을 찾기
max_diff_var = max(renewal_diffs, key=lambda x: renewal_diffs[x]['Difference'])
max_diff_info = renewal_diffs[max_diff_var]

max_diff_var, max_diff_info
```

분석 결과, 'EmploymentStatus' 변수가 보험금 청구가 1000 이상인 고객들 사이에서 갱신하는 비율과 갱신하지 않는 비율의 차이가 가장 큰 변수로 도출되었습니다. 특히, 'Unemployed' (실업 상태인) 고객들에서 이 차이가 가장 크며, 그 차이는 약 94.52% 입니다.

이 결과를 바탕으로, 실업 상태인 고객들이 보험 갱신을 상대적으로 적게 하는 경향이 있음을 알 수 있습니다. 이는 이익 증가(손실 감소)를 위한 전략을 수립할 때 중요한 고려 사항이 될 수 있습니다. 예를 들어, 실업 상태인 고객들에게 보험 갱신을 장려하기 위한 특별한 프로그램이나 혜택을 제공하는 것이 도움이 될 수 있습니다.
